```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
plt.rc('animation', html='html5')
```

```python
def func(x): return 1/(1+np.exp(x)) * np.sin(x)
```

```python
sz = 50
```

```python
x = sorted(2 * ((np.pi - (-np.pi)) * np.random.random_sample(size=sz) + (-np.pi)))
```

```python
y = func(x) + 0.2 * np.random.random_sample(size=sz)
```

```python
plt.scatter(x, y)
```

```python
steps = sz

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x, y, c='orange', label='Ground truth')
ax.set_ylim(-1, 2)
ax.set_title('Animated', fontsize=16)
line, = ax.plot([], [], label='Model')
ax.legend(loc='lower right')
title = fig.suptitle(f'Step: 0/{steps}', fontsize=22)
plt.close()

def animate(i):
    xi, yi = x[:i], y[:i]
    line.set_data(xi, yi)
    title.set_text(f'Step: {i+1}/{steps}')
    return line, title
```

```python
animation.FuncAnimation(fig, animate, range(steps+1), interval=40)
```
